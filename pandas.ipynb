```python
#1.most expensive car
def most_expensive_car():
    expensive_car =df. loc[df["CostPrice"]. idxmax()]
    return expensive_car
most_expensive_car()
```

```python
#2.getting the cheapest car
def cheapest_car():
    cheapest_car = df. loc[df["CostPrice"]. idxmin()]
    return cheapest_car
cheapest_car()
```
